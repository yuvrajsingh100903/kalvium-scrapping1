```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://results.eci.gov.in/PcResultGenJune2024/index.htm'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print (soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table',class_ ='table')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
world_table_titles = world_table_titles[0:4]
```

```python
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns = world_table_titles)
df
```

```python
table.find_all('tr')
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:43]:
    row_data_list = []  # Initialize a list to store row data
    for cell in row.find_all('td'):
        row_data_list.append(cell.text.strip()) # Extract and append text from each cell

    print(row_data_list)

    if len(row_data_list) == len(df.columns):
        length = len(df)
        df.loc[length] = row_data_list
```

```python
df
```

```python
from google.colab import files

df.to_csv('df.csv',index=False)
files.download('df.csv')
```

```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(12, 6))

# Bar plot for the number of seats won by each party
plt.barh(df['Party'], df['Total'], color='skyblue')
plt.xlabel('Number of Seats')
plt.ylabel('Party')
plt.title('Number of Seats Won by Each Party')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
